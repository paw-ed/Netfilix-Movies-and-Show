```python
sns.countplot(netfilix, x="type", palette="inferno")
plt.title("type")
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.title('tÃ¼r')
```

```python
ax = sns.countplot(data=netfilix, x="rating", palette="plasma")  # Create the countplot and assign it to ax
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right") # Rotate xticklabels
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title("rating")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="rating",hue="type",data=netfilix)
plt.title("tÃ¼r ve rating")
plt.show()
```

```python
Labels=["Movie","TV Show"]
size= netfilix["type"].value_counts()
cmap = plt.get_cmap('Wistia')
colors = cmap(np.linspace(0, 1, 2))
explode=[0,0.1]
plt.rcParams["figure.figsize"]=[7,7]
plt.pie(size,labels=Labels,colors=colors,explode=explode,shadow=True,startangle=90) # Fixed: Labels instead of labels
plt.title("tÃ¼r gÃ¶rÃ¼nÃ¼mÃ¼", fontsize=25)
plt.legend()
plt.show()
```

```python
netfilix["rating"].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(15,15))
plt.show()
```
